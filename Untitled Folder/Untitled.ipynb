```python
query = '''
SELECT 
	room_type, 
	toFloat32OrNull(replaceRegexpAll(price, '[$,]', '')) as price,
	toFloat32OrNull(review_scores_rating) as review_scores_rating
FROM 
    ldn_listings
WHERE review_scores_rating != NAN 
ORDER BY id
LIMIT 1000
'''
df = ph.read_clickhouse(query, connection=connection)
```

```python
query= ''' 
SELECT 
	DISTINCT host_id, 
	host_verifications
FROM 
    ldn_listings
WHERE experiences_offered != 'none'
'''
df = ph.read_clickhouse(query, connection=connection)
```

```python
df
```

```python
def create_list_host_verifications(list_expration):
    pattern = re.compile(r'[a-z_0-9]{1,30}')
    result = re.findall(pattern, list_expration.lower())
    return result
```

```python
def create_list_host_verifications_1(list_expration):
    list_expration = list_expration.strip("[] ").replace("'", '').replace(" ", '').split(",")
    return list_expration
```

```python
list_1 = "['email ', ' phone', 'facebook', 'reviews', 'offline_government_id', 'selfie', 'government_id']"
create_list_host_verifications_1(list_1)
```

```python
df_verifications_list = df
```

```python
df_verifications_list.host_verifications = df_verifications_list.\
    host_verifications\
    .apply(lambda x: create_list_host_verifications(x))
```

```python
df_verifications_list
```

```python
exploded_df = df_verifications_list.explode("host_verifications")
```

```python
exploded_df
```

```python
exploded_df.groupby('host_verifications').agg({'host_id':'count'})
```

```python
query = '''
SELECT 
	neighbourhood_cleansed, 
	experiences_offered,
	count(*) AS experiences_count
FROM 
    ldn_listings
WHERE experiences_offered != 'none'
GROUP BY 
	neighbourhood_cleansed, experiences_offered
ORDER BY 
	experiences_count DESC
limit 100'''

df = ph.read_clickhouse(query, connection=connection)
```

```python
df
```

```python
df_to_plot_heatmap = pd.pivot(df, index='neighbourhood_cleansed', 
                              columns='experiences_offered', 
                              values='experiences_count')
```

```python
cmap=sns.cubehelix_palette(as_cmap=True)
sns.heatmap(df_to_plot_heatmap, cmap=cmap) 
```

```python
query = '''SELECT 
	room_type, 
	 toFloat32OrNull(replaceRegexpAll(price, '[$,]', '')) as price_corr
FROM 
    ldn_listings
WHERE 
	experiences_offered != 'none'
ORDER BY 
	price_corr DESC
'''
df = ph.read_clickhouse(query, connection=connection)
```

```python
df
```

```python
df['loq_price'] = np.log(df.price_corr.values)
```

```python
df
```

```python
sns.set_style("whitegrid")
sns.histplot(data=df, x="price_corr", hue="room_type", kde=False)
```

```python
sns.displot(data=df, x="loq_price", hue="room_type")
```

```python
query='''
SELECT 
	room_type, 
	toFloat32OrNull(replaceRegexpAll(price, '[$,]', '')) as price_corr,
	first_review
FROM 
    ldn_listings
WHERE 
	first_review > '2010-01-02'
LIMIT 1000
'''
df = ph.read_clickhouse(query, connection=connection)
```

```python
df['first_review_year'] = df.first_review.apply(lambda x: x[:4])
```

```python
df_avg_price = df.groupby(['first_review_year', 'room_type'], as_index=False)\
    .agg({'price_corr':'mean'})\
    .sort_values('first_review_year')
```

```python
df_avg_price
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=df_avg_price, x='first_review_year', y='price_corr', hue='room_type')
```
