```python
import pandas as pd
import os
import datetime as dt
```

Объедините сэмплы данных за нужные месяцы в один общий датасет bikes. Сделайте преобразование по дням для каждой группы пользователей (usertype), затем выберите дни, в которые число аренд, сделанных customers, было больше, чем у subscribers.

```python
col = ['trip_id', 'start_time', 'end_time', 'bikeid', 'tripduration',
       'from_station_id', 'from_station_name', 'to_station_id',
       'to_station_name', 'usertype', 'gender', 'birthyear']
bikes = pd.DataFrame(columns=col)
row_len = 0

for item in os.listdir():
    if item[-3:] == 'csv':
        new_df = pd.read_csv(item, parse_dates=['start_time', 'end_time'])
        bikes = bikes.append(new_df, ignore_index=True)
        row_len += new_df.shape[0]
```

```python
if row_len == bikes.shape[0]:
    print('OK')
```

```python
bikes.head()
```

```python
bikes = bikes.sort_values(by='start_time')
bikes = bikes.set_index('start_time')
```

```python
bikes
```

```python
bikes_resample = bikes.groupby('usertype', as_index=False).resample(rule='D').size().T.reset_index()
bikes_resample.head()
```

выберите дни, в которые число аренд, сделанных customers, было больше, чем у subscribers.

```python
bikes_resample[bikes_resample[0]>bikes_resample[1]]
```

Еще один плюс использования дат в качестве индексов – возможность выбрать данные за интересующий нас промежуток времени. В переменную bikes_summer сохраните наблюдения с 1 июня по 31 августа. Затем запишите в top_destination наиболее популярный пункт назначения (его название). Агрегируйте данные по дням и определите, в какой день в полученный пункт (top_destination) было совершено меньше всего поездок. Дату сохраните в bad_day, отформатировав timestamp с помощью .strftime('%Y-%m-%d').

```python
bikes_summer = bikes.loc['2018-06-01':'2018-08-31'] #записываю даннные о летних деньках в датафрейм
bikes_summer.head()
```

```python
#Считаю количество поездок для каждого пункта назначения и ищу максимальное значение
destination_count = bikes_summer\
    .groupby('to_station_name', as_index=False)\
    .agg({'bikeid':'count'})
#1 способ
top_destination = destination_count.to_station_name.loc[destination_count.bikeid.idxmax()]
top_destination
```

```python
# 2 способ
destination_count.sort_values(by='bikeid', ascending=False).head(1)
```

```python
bad_day = bikes_summer.resample(rule='D').size().idxmin().strftime('%Y-%m-%d')
```

Куда больше всего ездят на выходных? Туда же, куда и в будние дни, или в другие пункты назначения?

Используя данные за период с 1 июня по 31 августа, выберите верные утверждения.

```python
bikes_summer = bikes.loc['2018-06-01':'2018-08-31']
```

```python
bikes_summer_to_week = bikes_summer.reset_index()
bikes_summer_to_week['weekday_name'] = bikes_summer_to_week.start_time.dt.day_name()
```

```python
bikes_summer_to_week.head()
```

```python
bikes_summer_to_week_agg = bikes_summer_to_week\
    .groupby(['weekday_name', 'to_station_name'], as_index=False)\
    .agg({'from_station_name':'count'})
bikes_summer_to_week_agg.head()
```

Наиболее популярный пункт назначения в субботу – Streeter Dr & Grand Ave

ответ - да

```python
saturday_max_idx = bikes_summer_to_week_agg[bikes_summer_to_week_agg.weekday_name == 'Saturday'].from_station_name.idxmax()
bikes_summer_to_week_agg.to_station_name.loc[saturday_max_idx] 
```

Второй по числу поездок в субботу – Lake Shore Dr & North Blvd

ответ - да

```python
bikes_summer_to_week_agg[bikes_summer_to_week_agg.weekday_name == 'Saturday'].sort_values(by='from_station_name').tail(2)
```

В будние дни больше всего поездок совершается до Streeter Dr & Grand Ave

ответ - да

```python
bikes_summer_to_week['weekend'] = bikes_summer_to_week.weekday_name.isin(['Saturday', 'Sunday'])
```

```python
bikes_to_week_agg = bikes_summer_to_week\
    .groupby(['weekend', 'to_station_name'], as_index=False)\
    .agg({'usertype':'count'})
bikes_to_week_agg
```
