Для начала, возьмем данные только за Q1, записать в переменную bikes_Q1. Перед тем как сделать .resample(), нужно немного подготовить данные. Поместите колонку start_time в качестве индексов и сохраните изменения в исходный датасет. Предварительно проверьте тип переменной, и приведите её к правильному, если необходимо.



**В качестве индексов датафрейма обязательно нужно использовать колонку формата DateTime, отсортированную в правильном порядке. Поэтому предварительно всегда следует проверить правильность типа данных и, если требуется, привести его к правильному с помощью pd.to_datetime().

```python
path_to_file = 'bikes_q1_sample.csv'
bikes_Q1 = pd.read_csv(path_to_file)
```

```python
bikes_Q1.head()
```

```python
bikes_Q1.start_time = pd.to_datetime(bikes_Q1.start_time)
bikes_Q1.end_time = pd.to_datetime(bikes_Q1.end_time)
```

Данные выгружены за 1 квартал

```python
bikes_Q1 = bikes_Q1.sort_values(by='start_time')
```

```python
bikes_Q1 = bikes_Q1.set_index('start_time')
```

```python
bikes_Q1
```

Необходимо найти максимальное число аренд за день 

```python
bikes_Q1_count_rentals_per_day = bikes_Q1.resample(rule='D').size()
bikes_Q1_count_rentals_per_day.idxmax()
```

Очень быстрый и легкий способ обрабатывать данные по временным промежуткам
